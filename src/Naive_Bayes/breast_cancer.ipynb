```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = '../../data/Prac1/breast-cancer-wisconsin.data'
columns = ['sample_code', 'clump_thickness', 'cell_size', 'cell_shape', 'marginal_adhesion', 'single_cell_size', 'bare_nuclei', 'bland_chromatin', 'normal_nucleoli', 'mitoses', 'class']
df = pd.read_csv(file_path, header=None, names=columns)
df
```

```python
df.describe()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
df['bare_nuclei'] = pd.to_numeric(df['bare_nuclei'], errors='coerce')
```

```python
df['bare_nuclei'].fillna(df['bare_nuclei'].median(), inplace=True)
```

```python
df_2 = df[df['class'] == 2]
df_2.shape
```

```python
df_4 = df[df['class'] == 4]
df_4.shape
```

```python
random_state = 18

test_set = pd.concat([df_2.sample(n=80, random_state=random_state), df_4.sample(n=40, random_state=random_state)])
test_set 
```

```python
train_set = pd.concat([df_2.drop(df_2.sample(n=80, random_state=random_state).index), df_4.drop(df_4.sample(n=40, random_state=random_state).index)])
train_set
```

```python
for column in columns[1:]:
    print(train_set[column].unique())
```

```python
X_train = train_set.drop('class', axis='columns')
y_train = train_set['class']
```

```python
X_test = test_set.drop('class', axis='columns')
y_test = test_set['class']
```

```python
X_train.info()
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred, average='macro')
```

```python
recall_score(y_test, y_pred, average='macro')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

**Lớp 2:**

* Precision = 0.89 → Trong số các mẫu dự đoán là "2", có 89% thực sự là "2".
* Recall = 0.95 → 95% số mẫu thực sự là "2" đã được mô hình phát hiện đúng.
* F1-score = 0.92 → Hiệu suất tổng thể tốt.

**Lớp 4:**

* Precision = 0.89 → Trong số các mẫu dự đoán là "4", có 89% thực sự là "4".
* Recall = 0.78 → 78% số mẫu thực sự là "4" đã được mô hình phát hiện đúng.
* F1-score = 0.83 → Tổng thể khá.

### Các chỉ số tổng hợp:

* Accuracy (Độ chính xác tổng thể) = 0.89 (~89%)
    * Tỷ lệ dự đoán đúng trên tổng số mẫu.​
 
* Macro avg (Trung bình không cân bằng) = 0.89, 0.86, 0.87
    * Trung bình cộng của Precision, Recall và F1-score của từng lớp.
* Weighted avg (Trung bình có cân nhắc số lượng mẫu) = 0.89, 0.89, 0.89
    * Giống macro avg nhưng có tính đến số lượng mẫu (support).
