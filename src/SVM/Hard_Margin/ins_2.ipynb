```python
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
file_path = "../../data/SVM/sonar.all-data.csv"

data = pd.read_csv(file_path, header=None)

data.head()
```

```python
X = data.iloc[:, :60]
y = data.iloc[:, -1]
```

```python
y = y.map({"R": -1, "M": 1})
```

```python
clf = SVC(kernel='linear', C = 1e6)

clf.fit(X, y)

y_pred = clf.predict(X)

print("Accuracy:", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred))
print("Recall:", recall_score(y, y_pred))
```

```python
clf = SVC(kernel='linear', C = 3.5)

clf.fit(X, y)

y_pred = clf.predict(X)

print("Accuracy:", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred))
print("Recall:", recall_score(y, y_pred))
```

### Nhận xét:

* Trường hợp C >= 10^5:
    * Accuracy = 1 
    * Precision = 1
    * Recall = 1

    => Mô hình có độ chính xác tuyệt đối, nhưng khả năng overfitting rất cao.

* Trường hợp C trong [0.5, 5], chọn C = 3.5:
    * Accuracy = 0.8509615384615384
    * Precision = 0.8333333333333334
    * Recall = 0.9009009009009009
    
    => Mô hình đạt hiệu suất tốt và cân bằng, phù hợp cho tổng quát hóa.

```python
parameters = {'kernel': ['linear'], 'C':[1e5, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]}
svc = SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(X, y)
```
