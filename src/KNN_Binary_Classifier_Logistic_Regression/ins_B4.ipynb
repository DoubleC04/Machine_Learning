```python
import pandas as pd
import numpy as np
import time
```

```python
file_path = '../../data/Prac3/banking.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
df.head()
```

```python
df1 = df.copy()
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_encoder = LabelEncoder()
df1['month'] = le_encoder.fit_transform(df1['month'])
df1.head()
```

```python
df1['day_of_week'] = le_encoder.fit_transform(df1['day_of_week'])
df1.head()
```

```python
df1['default'] = le_encoder.fit_transform(df1['default'])
df1.head()
```

```python
df1['housing'] = le_encoder.fit_transform(df1['housing'])
df1.head()
```

```python
df1['loan'] = le_encoder.fit_transform(df1['loan'])
df1.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
```

```python
def ohe_encoder(ohe, cols, df):
    df_encoded = df.copy()
    for col in cols:
        encoded_data = ohe.fit_transform(df_encoded[[col]])
        
        encoded_df = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out([col]))
        
        df_encoded = pd.concat([df_encoded, encoded_df], axis='columns')
        df_encoded.drop(col, axis='columns', inplace=True)
    
    return df_encoded
```

```python
df2 = ohe_encoder(ohe, ['job', 'marital', 'education', 'contact', 'poutcome'], df1)
```

```python
df2.head()
```

```python
df2['pdays'] = df2['pdays'].apply(lambda x: 0 if x == -1 else 1)
```

```python
df2.info()
```

```python
X = df2.drop('y', axis='columns')
y = df2['y']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
logR = LogisticRegression()
start_time = time.time()
logR.fit(X_train, y_train)

y_pred_logR = logR.predict(X_test)
run_time = time.time() - start_time

print(f"Time: {run_time:.6f} s")
print(accuracy_score(y_test, y_pred_logR))
print(precision_score(y_test, y_pred_logR))
print(accuracy_score(y_test, y_pred_logR))
print(f1_score(y_test, y_pred_logR))
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
start_time = time.time()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)
run_time = time.time() - start_time

print(f"Time: {run_time:.6f} s")
print(accuracy_score(y_test, y_pred_nb))
print(precision_score(y_test, y_pred_nb))
print(accuracy_score(y_test, y_pred_nb))
print(f1_score(y_test, y_pred_nb))
```
