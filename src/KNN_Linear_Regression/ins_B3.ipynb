```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '../../data/Prac2/SAT_GPA.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
corr_values = df.corr()['GPA']

plt.figure(figsize=(8, 6))
sns.heatmap(corr_values.to_frame(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()
```

```python
X = df['SAT']
y = df['GPA']
```

```python
X_train = X[:64]
y_train = y[:64]

X_test = X[64:]
y_test = y[64:]
```

```python
from sklearn.linear_model import LinearRegression

X_train = X_train.to_frame()

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
plt.scatter(X, y)

theta_1 = lr.coef_
theta_0 = lr.intercept_

yhat = theta_1 * X + theta_0

plt.plot(X, yhat, label='regression line', c='gold', lw=3)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_test = X_test.to_frame()
y_pred = lr.predict(X_test)

print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
